```python
def voteCount(df,year):
    house = df[df["year"]==year]
    dfp = pd.pivot_table(data = house,
                                                index = 'state',
                                                columns = 'party',
                                                values = 'candidatevotes',
                           aggfunc = np.sum)
    dfp = dfp.reset_index().fillna(0)

    try:
        try:
            dfp["DEMOCRAT"] = dfp['DEMOCRAT'] + dfp['DEMOCRATIC-FARMER-LABOR']
            dfp["DEMOCRAT"] = dfp['DEMOCRAT'] + dfp['DEMOCRATIC-NPL']
            dfp = dfp.drop(['DEMOCRATIC-NPL','DEMOCRATIC-FARMER-LABOR'],axis=1)
        except:
            try:
                dfp["DEMOCRAT"] = dfp['DEMOCRAT'] + dfp['DEMOCRATIC-FARMER-LABOR']
                dfp = dfp.drop(['DEMOCRATIC-FARMER-LABOR'],axis=1)
            except:
                pass
    except:
        pass
    dfp['Total'] = dfp.sum(1)
    dfp['OTHER'] = dfp['Total'] - (dfp['DEMOCRAT']+dfp['REPUBLICAN'])
    ndf = dfp[['state','DEMOCRAT','REPUBLICAN','OTHER','Total']]
    ndf['State Name'] = ndf['state'].apply(lambda x: x.lower().lstrip().rstrip())
    ndf['demShare'] = ndf['DEMOCRAT']/ndf['Total']
    ndf['gopShare'] = ndf['REPUBLICAN']/ndf['Total']
    ndf['indShare'] = ndf['OTHER']/ndf['Total']
    ndf['margin'] = ndf['gopShare'] - ndf['demShare']
    
    return ndf
```

```python
votes18 = voteCount(df,2018)
votes18.to_csv('2018labels.csv')
votes18
```
